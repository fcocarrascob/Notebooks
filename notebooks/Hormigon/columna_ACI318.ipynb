```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Leer desde el portapapeles, con tabulaciones como separadores y coma como decimal, sin encabezados
df = pd.read_clipboard(sep='\t', decimal=',', header=None)

#Pre procesar datos cambiar "," por "." y convertir a float
df = df.apply(lambda col: col.map(lambda x: pd.to_numeric(str(x).replace(',', '.'), errors='coerce')))
```

```python
#Grafico P-M3
#Curva 1 Angulo 0
curva1_P = df[1][2:]
curva1_M = df[3][2:]

#Curva 13 Angulo 180
curva13_P = df[37][2:]
curva13_M = df[39][2:]
```

```python
plt.plot(curva1_M, curva1_P, label='Curva 1 Angulo 0')
plt.plot(curva13_M, curva13_P, label='Curva 13 Angulo 180')

#Nombre de ejes
plt.xlabel('Momento $M_n$ (tonf-m)')
plt.ylabel('Carga $P_n$ (tonf)')

#Titulo del grafico
plt.title('Grafico P-M3')

#cuadricula
plt.grid()

#ajustar valores de los ejes

from matplotlib.ticker import MaxNLocator, AutoMinorLocator

ax = plt.gca()
ax.relim(); ax.autoscale_view()
ax.xaxis.set_major_locator(MaxNLocator(nbins=8))
ax.yaxis.set_major_locator(MaxNLocator(nbins=8))
ax.xaxis.set_minor_locator(AutoMinorLocator(4))
ax.yaxis.set_minor_locator(AutoMinorLocator(4))
ax.grid(which='major', alpha=0.7)
ax.grid(which='minor', alpha=0.3, linestyle='--')
plt.tight_layout()

#mostrar grafico
plt.show()
```
