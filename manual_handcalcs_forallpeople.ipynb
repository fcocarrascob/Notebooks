## 3. `handcalcs` — Importar y Configurar

`handcalcs` convierte automáticamente expresiones de Python en ecuaciones LaTeX renderizadas dentro del notebook. Funciona como una **extensión mágica** de Jupyter.

### Conceptos clave

| Comando | Descripción |
|---|---|
| `%load_ext handcalcs.render` | Carga la extensión una sola vez por sesión |
| `%%render` | Magic de celda: renderiza **todas** las asignaciones como LaTeX |
| `%%render symbolic` | Solo muestra la forma simbólica (sin sustitución numérica) |
| `%%render params` | Muestra una tabla compacta de parámetros |
| `%%render long` | Muestra la sustitución completa en formato extendido |

### ¿Cómo funciona?

Cuando usas `%%render`, `handcalcs` inspecciona el código de la celda, extrae cada asignación (`variable = expresión`) y genera la representación LaTeX:

$$A = b \cdot h = 5 \cdot 3 = 15$$


````python
import handcalcs.render          # importar el módulo
%load_ext handcalcs.render        # cargar la extensión mágica en Jupyter

print("handcalcs cargado correctamente ✓")
````

## 4. Uso Básico: Expresiones Simples

La magia `%%render` (dos signos `%`) se coloca **al inicio de la celda**. Cada línea que contenga una asignación (`=`) se renderiza como una ecuación LaTeX.

**Reglas básicas:**
- Las variables deben estar definidas antes o dentro de la misma celda
- Se admiten operaciones: `+`, `-`, `*`, `/`, `**` (potencia), `sqrt()`, funciones de `math`
- Los nombres de variables con subscript usan `_`: `F_y` → $F_y$, `sigma_long` → $\sigma_{long}$
- Los nombres con superscript usan `__`: `x__2` → $x^{2}$


````python
%%render
# Cálculo de área y perímetro de un rectángulo
b = 5       # base (m)
h = 3       # altura (m)

A = b * h
P = 2 * (b + h)
````

````python
%%render
# Ejemplo con fórmulas más complejas — tensión axial
import math

N = 150000     # Carga axial (N)
d = 0.05       # Diámetro sección circular (m)

A_circ = math.pi * d**2 / 4
sigma = N / A_circ
````

## 5. Comentarios y Modo Simbólico

### Comentarios como encabezados de sección (`##`)

Dentro de una celda `%%render`, los comentarios con **doble almohadilla** (`## Texto`) se convierten en **encabezados** de sección dentro del bloque renderizado. Un solo `#` es un comentario normal de Python y es ignorado.

### Modo simbólico (`%%render symbolic`)

El modo simbólico muestra **solo la fórmula algebraica**, sin sustituir los valores numéricos. Es útil para documentar la metodología de cálculo o para presentaciones.


````python
%%render
## Propiedades de la sección
b_s = 0.30      # Ancho de viga (m)
h_s = 0.60      # Altura de viga (m)
A_s = b_s * h_s

## Momento de inercia
I_x = (b_s * h_s**3) / 12
````

````python
%%render symbolic
# Modo simbólico: solo muestra la fórmula, sin números
b_sym = 0.30
h_sym = 0.60
A_sym = b_sym * h_sym
I_sym = (b_sym * h_sym**3) / 12
````

## 6. Modos de Renderizado: `params` y `long`

`handcalcs` ofrece distintos modos de visualización según el contexto del cálculo:

| Modo | Comando | Uso recomendado |
|---|---|---|
| **Normal** | `%%render` | Renderizado estándar: simbólico + sustitución + resultado |
| **Parámetros** | `%%render params` | Tabla compacta de valores de entrada (sin fórmulas) |
| **Formato extendido** | `%%render long` | Muestra cada operación en línea separada (fórmulas grandes) |

### Etiquetas en línea

También puedes controlar el renderizado de variables individuales usando etiquetas en los comentarios:
- `# Parameters` → trata la línea como parámetro de entrada
- `# Result` → destaca el resultado final

## 7. `forallpeople` — Importar y Configurar

`forallpeople` permite trabajar con **cantidades físicas con unidades** directamente en Python, siguiendo el Sistema Internacional (SI). Las unidades se propagan automáticamente en los cálculos.

### Entornos disponibles

Al llamar `si.environment(nombre)`, se carga un conjunto predefinido de unidades:

| Entorno | Unidades incluidas | Uso típico |
|---|---|---|
| `'default'` | Unidades SI base: m, kg, s, A, K, mol, cd | Física general |
| `'structural'` | kN, m, kPa, MPa, kN·m, ton, cm, mm | Ingeniería estructural |
| `'mechanical'` | N, mm, MPa, J, W | Ingeniería mecánica |
| `'civil'` | m, kN, kPa, ton | Obras civiles |

### ¿Cómo funciona?

Multiplicas un número por una unidad del entorno cargado y obtienes una cantidad física. `forallpeople` gestiona dimensiones, conversiones y renderizado automáticamente:

```python
fuerza = 10 * si.kN   # → 10.00 kN
area   = 0.05 * si.m**2
```


````python
import forallpeople as si

# Cargar el entorno de ingeniería estructural
# Esto inyecta unidades como kN, m, MPa, kPa, mm, cm, etc.
si.environment('structural', top_level=True)

print("forallpeople cargado con entorno 'structural' ✓")
print(f"Unidades disponibles directamente: kN, m, kPa, MPa, mm, cm, ton, kN·m")
````

## 8. Operaciones Básicas con Unidades

Una vez cargado el entorno, puedes crear cantidades físicas multiplicando un número por la unidad. Las operaciones aritméticas entre cantidades propagan las unidades automáticamente.

**Reglas importantes:**
- Suma y resta: solo entre cantidades con **las mismas dimensiones**
- Multiplicación / división: las unidades se combinan → p.ej. `kN / m²` = `kPa`
- Potencias: `longitud ** 2` → unidad cuadrada


````python
# Definir cantidades físicas con unidades
fuerza  = 150 * kN       # 150 kN
largo   = 6.0 * m        # 6 m
ancho   = 0.30 * m       # 30 cm
alto    = 0.60 * m       # 60 cm

# Operaciones — las unidades se propagan automáticamente
area_secc = ancho * alto          # m² 
momento   = fuerza * largo / 4   # kN·m (viga simplemente apoyada, carga puntual centro)

print(f"Fuerza:           {fuerza}")
print(f"Largo:            {largo}")
print(f"Área de sección:  {area_secc}")
print(f"Momento flector:  {momento}")
````

## 9. Conversión de Unidades

`forallpeople` permite convertir entre unidades compatibles. Los métodos más comunes son:

| Método | Ejemplo | |
|---|---|---|
| `.to(unidad)` | `fuerza.to(N)` | Convierte a la unidad indicada |
| Dividir por unidad | `fuerza / kN` | Extrae el valor numérico en esa unidad |
| `.prefix(prefijo)` | `longitud.prefix('m')` | Cambia el prefijo (k, M, m, μ…) |

> **Importante:** no se puede convertir entre unidades de dimensiones diferentes (p.ej. `m` a `kN`); hacerlo lanzará un error dimensional.


````python
# Conversiones de unidades
f_kN  = 250 * kN
l_m   = 3500 * mm        # 3500 mm = 3.5 m
p_MPa = 25 * MPa

# Conversión de kN a N
f_N = f_kN.to(N)
print(f"Fuerza en kN: {f_kN}")
print(f"Fuerza en N:  {f_N}")

# mm a m
l_m_conv = l_m.to(m)
print(f"\nLongitud en mm: {l_m}")
print(f"Longitud en m:  {l_m_conv}")

# MPa a kPa
p_kPa = p_MPa.to(kPa)
print(f"\nPresión en MPa: {p_MPa}")
print(f"Presión en kPa: {p_kPa}")

# Extraer solo el valor numérico (adimensional) en una unidad dada
valor_numerico = float(f_kN / kN)
print(f"\nValor numérico de la fuerza en kN: {valor_numerico}")
````

## 10. Uso Combinado: `handcalcs` + `forallpeople`

La combinación de ambas librerías permite escribir **memorias de cálculo profesionales** directamente en Jupyter: las variables llevan unidades (gracias a `forallpeople`) y las ecuaciones se renderizan como LaTeX legible (gracias a `handcalcs`).

### Ejemplo: Verificación de tensión por flexión en viga rectangular

Dado:
- Viga de sección rectangular, simplemente apoyada
- Carga puntual en el centro del vano

Comprobar que la tensión de trabajo $\sigma$ no supera la tensión admisible $f_y$.

$$\sigma = \frac{M \cdot (h/2)}{I} \leq f_{adm}$$


````python
%%render params
## Datos de entrada
P_carga = 200 * kN         # Carga puntual aplicada
L_viga  = 6.0 * m          # Luz de la viga
b_viga  = 300 * mm         # Ancho de la sección
h_viga  = 600 * mm         # Altura de la sección
f_adm   = 120 * MPa        # Tensión admisible del material
````

````python
%%render
## Propiedades geométricas de la sección
A_secc = b_viga * h_viga
I_secc = (b_viga * h_viga**3) / 12
W_secc = I_secc / (h_viga / 2)

## Esfuerzos internos (viga simplemente apoyada, carga puntual centro)
M_max = (P_carga * L_viga) / 4

## Verificación de tensión por flexión
sigma_max = M_max / W_secc
````

````python
# Verificación final — comparación de tensiones
sigma_val = float(sigma_max / MPa)
f_adm_val = float(f_adm / MPa)

estado = "CUMPLE ✓" if sigma_val <= f_adm_val else "NO CUMPLE ✗"

print(f"Tensión de trabajo:   {sigma_max:.3f}")
print(f"Tensión admisible:    {f_adm}")
print(f"Verificación:         {sigma_val:.2f} MPa ≤ {f_adm_val:.0f} MPa → {estado}")
````
