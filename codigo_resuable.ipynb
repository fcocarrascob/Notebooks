Solución: usar __wrapped__
Los decoradores de handcalcs preservan la función original en el atributo __wrapped__. Al llamarlo directamente, omites el renderizado:

```python
@handcalc(jupyter_display=True)
def Q_iw(y, h_liq, D):
    Q_iw = 0.866 * (1 - (y / h_liq)**2) * np.tanh(0.866 * D / h_liq)
    return Q_iw

# Calcular Q_iw en la base
Q_iw_max = Q_iw(0, h_liq.value, D.value)
```
